```python
movies['year'] = movies.title.str.extract('(\(\d\d\d\d\))', expand=False)
```

```python
movies['year'] = movies.year.str.extract('(\d\d\d\d)', expand=False)
```

```python
movies['title'] = movies.title.str.replace('(\(\d\d\d\d\))','')
```

```python
movies['title'] = movies['title'].apply(lambda x:x.strip())
movies.head()
```

```python
movies['genres'] = movies.genres.str.split('|')
```

```python
movies_WithGenres = movies.copy()
```

```python
for index,row in movies.iterrows():
    for genre in row['genres']:
        movies_WithGenres.at[index, genre] = 1
movies_WithGenres = movies_WithGenres.fillna(0)        
```

# USER INPUT

```python
user_input = [
       {'title':'GoldenEye', 'rating':1},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Tom and Huck", 'rating':5},
            {'title':'Sudden Death', 'rating':4.5},
    
]
```

```python
customer_input = pd.DataFrame(user_input)
```

```python
customer_input
```

```python
inputId = movies[movies['title'].isin(customer_input['title'].tolist())]
```

```python
inputId
```

```python
input_movies = pd.merge(inputId,customer_input)
```

```python
input_movies = input_movies.drop('genres',1).drop('year',1)
```

```python
input_movies
```

```python
user_input = movies_WithGenres[movies_WithGenres['movieId'].isin(input_movies['movieId'].tolist())]
```

```python
user_input = user_input.reset_index(drop=True)
```

```python
usergenretable = user_input.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
```

```python
usergenretable
```

```python
input_movies['rating']
```

```python
userprofile = usergenretable.transpose().dot(input_movies['rating'])
```

# user preferred Genres

```python
userprofile
```

```python
genretable = movies_WithGenres.set_index(movies_WithGenres['movieId'])
```

```python
genretable = genretable.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
```

```python
genretable.shape
```

```python
recommendationTable = ((genretable*userprofile).sum(axis=1))/(userprofile.sum())
```

```python
recommendationTable.head()
```

```python
recommendationTable = recommendationTable.sort_values(ascending=False)
recommendationTable.head()
```
